# Tuglet

```python
# Read the correct experiment
def get_experiment(
    oversampling_mode:str, oversampling_attribute:str, 
    metrics: list, fairness_metrics:list, demographics: list, combined_demographics:list, 
    filtering_experiments: list, filtering_columns: list):
    if oversampling_mode == 'baseline':
        root_name = 'no_os/tuglet/'.format(oversampling_attribute)
        regexes = ['no_os/(.*)/tuglet']
    else:
        root_name = 'oversample/{}_oversampling/{}'.format(oversampling_mode, oversampling_attribute)
        regexes = ['oversample/(.*)/tuglet/']

    labels = 'pass'
    
    settings = {
        'experiment': {
            'root_name': root_name,
            'labels': labels,
            'exclude': ['nothingtoexclude'],
            'regexes': regexes,
            'n_classes': 2
        },
        'data': {
            'metrics': metrics,
            'fairness_metrics': fairness_metrics,
            'demographics': demographics,
            'combined_demographics': combined_demographics
        },
        'csv': {
            'folder': '../experiments/oversample',
            'file': 'tuglet.tsv',
            'sorting': ['metric', 'baseline'],
            'baseline': 'no_os',
            'fixed': ['key', 'oversampling_model', 'oversampling_proportion', 'oversampling_attribute', 'metric'],
            'filtering_experiments': filtering_experiments,
            'filtering_columns': filtering_columns
        },
        'style': {
            'figsize_width': 'x',
            'figsize_height': 'x'
        },
        'boxplot': True,
        'fairness': True,
        'equal_odds': True,
        'fold': True,
        'colour': True,
        'drop_std': True,
        'overall': False,
        'filtering': True,
        'sort': True,
        'dump': True,
        'print': False,
        'show': False,
        'save': False,
        'display': True ,
        'combined': True,
        'barplot': True,
        'pairwise': False
    }
    
    plotter = InverseNonNestedPlotter(settings)
    plotter.test(settings)
    
    settings['overall'] = True
    settings['barplot'] = True
    settings['combined'] = True
    plotter = InverseNonNestedPlotter(settings)
    plotter.test(settings)
    
    
    return settings

def plot(oversampling_modes:list, oversampling_attribute:str, 
    metrics: list, fairness_metrics:list, demographics: list, combined_demographics:list, 
         filtering_experiments: list, filtering_columns: list):
    for om in oversampling_modes:
        settings = get_experiment(
            om, oversampling_attribute, 
            metrics, fairness_metrics, demographics, combined_demographics, filtering_experiments, filtering_columns
        )
    return settings
        
            
def display_all(settings):
    viewer = CSVViewer(settings)
    df = viewer.view_csv()
```

```python
# Choose parameters         
oversampling_modes = [
    'baseline', 
    'equal_balancing', 'major', 'minor'
]

metrics = [
    'roc', 'balanced_accuracy', 'tp', 'fp', 'fn'
]

fairness_metrics = [
    'roc', 'balanced_accuracy', 'tp', 'fp', 'fn'
]

demographics = [
    'gender', 'area', 'cluster'
]

oversampling_attributes = [
    'gender_area', 'cluster'
]

combined_demographics = ['gender', 'area']

filtering_experiments = []
filtering_columns = []

for oa in oversampling_attributes:
    oversampling_attribute = oa
    settings = plot(
        oversampling_modes,oversampling_attribute, metrics, fairness_metrics, demographics, combined_demographics, filtering_experiments, filtering_columns
    )
```

```python
filtering_experiments = {
    'oversampling_attribute': ['none', 'gender_area', 'gender', 'cluster', 'area'],
    'oversampling_proportion': ['minor', 'none']
}
filtering_columns = ['overall', 'gender_f', 'gender_m', 'area_t', 'area_r', 'cluster']
# filtering_columns = []
settings['csv']['filtering_columns'] = filtering_columns
settings['csv']['filtering_experiments'] = filtering_experiments
display_all(settings)
```
