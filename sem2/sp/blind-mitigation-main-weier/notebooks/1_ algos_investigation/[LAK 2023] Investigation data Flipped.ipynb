# Results

```python
result_path = './experiments/oversample/no_os/flipped/vinitra_ensemble_but_mirko/flipped_bilstm/pass/'
result_path += '2022_08_16_0/results/nonnested_cval_mfcbilstm_lx.pkl'
with open(result_path, 'rb') as fp:
    results = pickle.load(fp)
```

```python
results['demographics']
```

```python
genders = {}
y_pred = {}
y_proba = {}
y_tests = {}

rocs = {}
for i in range(10):
    y_pred[i] = results[i]['y_pred']
    y_proba[i] = results[i]['y_proba']
    genders[i] = [results['demographics']['gender'][idx] for idx in results[i]['test_index']]
    y_tests[i] = [results['y'][idx] for idx in results[i]['test_index']]
    
    fem_idx = [idx for idx in range(len(genders[i])) if genders[i][idx] == 'F']
    mal_idx = [idx for idx in range(len(genders[i])) if genders[i][idx] == 'M']
    
    fem_pred = [y_pred[i][idx] for idx in fem_idx]
    mal_pred = [y_pred[i][idx] for idx in mal_idx]
    
    fem_proba = [y_proba[i][idx][1] for idx in fem_idx]
    mal_proba = [y_proba[i][idx][1] for idx in mal_idx]
    
    fem_y = [y_tests[i][idx] for idx in fem_idx]
    mal_y = [y_tests[i][idx] for idx in mal_idx]
    
    
    print(fem_y)
    try:
        rocs[i] = {
            'general' : roc_auc_score(y_tests[i], [yp[1] for yp in y_proba[i]]),
            'female': roc_auc_score(fem_y, fem_proba),
            'male': roc_auc_score(mal_y, mal_proba)
        }
    except ValueError:
        print('oh no')
    print()
    
    
    
```

```python
rocs
```

```python
means = {
    'general': np.mean([rocs[fold]['general'] for fold in range(10) if fold in rocs]),
    'males': np.mean([rocs[fold]['male'] for fold in range(10) if fold in rocs]),
    'females': np.mean([rocs[fold]['female'] for fold in range(10) if fold in rocs])
}
```
